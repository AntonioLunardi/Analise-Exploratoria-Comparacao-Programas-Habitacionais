```python
# Atribuindo a base de dados a uma variável e exibindo-a
dados_brutos_MCMV = pd.read_csv(r'C:\Users\USER\Desktop\CDHU_MCMV\MCMV\MCMV.csv', sep=';', encoding = 'latin-1')
dados_brutos_MCMV.head()
```

```python
dados_brutos_MCMV.shape
```

```python
dados_brutos_MCMV.isna().value_counts() # Quantidade de valores faltantes
```

```python
# Excluindo as colunas que não serão utilizadas
dados_brutos_MCMV = dados_brutos_MCMV.drop(['UF', 'COD_IBGE', 'CAPITAIS', 'UF_1', 
                                            'UH_CONC', 'FID', 'the_geom',
                                            'Sub_FGTS', 'Sub_OGU', 'Percentual', 'IF'], axis = 1)
dados_brutos_MCMV.head()
```

```python
# Trocando o nome da variável e filtrando as cidades de estudo
dados_MCMV = dados_brutos_MCMV

dados_MCMV = dados_MCMV[(dados_MCMV['Municipio'] == 'Araraquara') | (dados_MCMV['Municipio'] == 'Bauru')
| (dados_MCMV['Municipio'] == 'RibeirÃ£o Preto') | (dados_MCMV['Municipio'] == 'SÃ£o JosÃ© dos Campos')
| (dados_MCMV['Municipio'] == 'SÃ£o Paulo')      | (dados_MCMV['Municipio'] == 'Registro')
| (dados_MCMV['Municipio'] == 'MarÃ­lia')        | (dados_MCMV['Municipio'] == 'Campinas')
| (dados_MCMV['Municipio'] == 'AraÃ§atuba')      | (dados_MCMV['Municipio'] == 'Franca')
| (dados_MCMV['Municipio'] == 'Itapeva')         | (dados_MCMV['Municipio'] == 'Presidente Prudente')
| (dados_MCMV['Municipio'] == 'Santos')          | (dados_MCMV['Municipio'] == 'Sorocaba')]

dados_MCMV['Municipio'].unique()
```

```python
a_trocar = {
    'RibeirÃ£o Preto': 'Ribeirão Preto',
    'SÃ£o Paulo': 'São Paulo',
    'Santos': 'Santos',
    'MarÃ­lia': 'Marília',
    'SÃ£o JosÃ© dos Campos': 'São José dos Campos',
    'AraÃ§atuba': 'Araçatuba',
    'Itapeva': 'Itapeva',
    'Bauru': 'Bauru',
    'Campinas': 'Campinas',
    'Franca': 'Franca',
    'Presidente Prudente': 'Presidente Prudente',
    'Sorocaba' : 'Sorocaba',
    'Registro' : 'Registro', 
    'Araraquara' : 'Araraquara'
}
dados_MCMV['Municipio'] = dados_MCMV['Municipio'].map(a_trocar)
dados_MCMV['Municipio'].unique()
```

```python
cidades = pd.DataFrame(data={'Cidade': dados_MCMV['Municipio'].unique()})
cidades.head()
```

```python
coordenadas_cidades = pd.DataFrame(data={'Longitude': dados_MCMV['XCOORD'].unique(), 'Latitude': dados_MCMV['YCOORD'].unique()})
coordenadas_cidades
```

```python
# Consertando o erro de conversão que foi realizado, no qual multiplicou-se por 10 o valor 
# de quase todos os campos de longitude e todos de latitude

for i in range(0, len(coordenadas_cidades)):
    if coordenadas_cidades['Longitude'][i]<-50:
        coordenadas_cidades['Longitude'][i] = coordenadas_cidades['Longitude'][i]/10
        
coordenadas_cidades['Latitude'] = coordenadas_cidades['Latitude']/10
coordenadas_cidades
```

```python
# Excluindo a última linha, que não diz respeito a nenhuma cidade da lista, provavelmente um erro de digitação
coordenadas_cidades = coordenadas_cidades.drop(index=14)

coordenadas_cidades
```

```python
# Criando data frame de cidades e coornenadas geográficas
df_coordenadas_cidades = pd.concat([cidades, coordenadas_cidades], axis=1)

df_coordenadas_cidades
```

```python
dados_MCMV.loc[dados_MCMV['XCOORD'] == -462.241] # Verificando quais as linhas com o registro problemático 
# com o valor de longitude da tabela original no df_coordenadas_cidades, a linha não precisará ser tratada
# em separado
```

```python
# Como as colunas das coordenadas serão posteriormente removidas da tabela e as coordenadas da cidade de Itapeva corretas
# já estão presentes 
```

```python
# Devido ao fato de haver muitos valores incorretos do MCMV, como pode-se observar, a tabela df_coordenadas_cidades será
# utilizada para informar as coordenadas das cidades das bases dos dois programas habitacionais através de chaves estrangeiras 
# no Power BI. Assim, as colunas de longitude e latitude do data frame MCMV serão excluídas

dados_MCMV = dados_MCMV.drop(columns=['XCOORD', 'YCOORD'])

dados_MCMV.head()
```

```python
# Criando colunas para análise

dados_MCMV['Concluídas não entregues'] = dados_MCMV['Concluidas'] - dados_MCMV['Entregues']
dados_MCMV['Não concluídas'] = dados_MCMV['UH'] - dados_MCMV['Concluidas']
```

```python
# Deixando em ordem alfabética para certificar a organização

df_coordenadas_cidades = df_coordenadas_cidades.sort_values('Cidade')

df_coordenadas_cidades.head()
```

```python
# Criando uma coluna de gasto MCMV por município

gastos_medios_casa_cidade = dados_MCMV['Valor'].groupby(dados_MCMV['Municipio']).sum()/dados_MCMV['UH'].groupby(dados_MCMV['Municipio']).sum()

# Realizando o arrendondamento dos valores ao criar o df para concatenação

gastos_medios_casa_cidade = pd.DataFrame({'Gasto médio por UH (MCMV)': gastos_medios_casa_cidade.round()}) 

# Resetando o index para fazer a junção

gastos_medios_casa_cidade = gastos_medios_casa_cidade.reset_index()

gastos_medios_casa_cidade.head()
```

```python
# Unindo as tabelas

df_coordenadas_cidades = pd.concat([df_coordenadas_cidades, gastos_medios_casa_cidade['Gasto médio por UH (MCMV)']], axis=1)

df_coordenadas_cidades
```

```python
# Como salvo anteriormente, havia por alguma razão certas células com um número maior de casas decimais

# Os seguintes comandos certificam que não serão gerados registros com número de casas demimais maiores
# que o padronizado para latitude e longitude (4 casas)

df_coordenadas_cidades['Longitude'] = round(df_coordenadas_cidades['Longitude'], 4)
df_coordenadas_cidades['Latitude'] = round(df_coordenadas_cidades['Latitude'], 4)
```

```python
# Para a visualização em Power BI do relatório será utilizada a tabela df_coordenadas_cidades, de modo
# a se relacionar com a base dados_CDHU com uma chave estrangeira de cidade, para obter as respectivas
# coordenadas. Dessa forma será possível incluir a tabela na visualização do widget de mapa dinâmico.
```
